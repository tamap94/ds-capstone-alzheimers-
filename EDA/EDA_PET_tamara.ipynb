```python
# Use the function from Valentin 
def get_csvdata_ADNI():
    df = pd.read_csv("../Data/Tables/ADNI_2_8_24_2022.csv").sort_values(["Subject","Description"])
    df.rename(columns={"Subject":"ID"}, inplace=True)
    df= df[(df["Description"] == "Talairach Warped")]
    image_IDs = []
    for i in df["ID"].unique():
        image_IDs.append(df[df["ID"]==i]["Image Data ID"].iloc[0])
    df= df.loc[df["Image Data ID"].isin(image_IDs)]
    #logger.info("ADNI-csv loaded")
    return df
```

```python
df = get_csvdata_ADNI()
df.head()
```
