```python
#Import packages 
import graphviz 
```

```python
#Define the FlowChart architecture 

dot = graphviz.Digraph('Workflow')

dot.attr('node', shape='box', fixedsize='true', width='1.4')
dot.node('a', 'Load data')
dot.node('c', 'Train-Test split')
dot.node('e', 'Make predictions')
dot.node('f', 'Analyze output')

dot.attr('node', shape='box', style='filled', color='lightgrey', fixedsize='true', width='1.8')
dot.node('b', label='Image pre-processing')
dot.node('d', 'Modeling')

```

```python
#Generate the Chart

dot.edges(['ab', 'bc', 'cd', 'de', 'ef'])
dot.attr(fontsize='20')

dot.graph_attr['rankdir']='LR'
dot.edge_attr.update(arrowhead='normal', arrowsize='2')

dot
```

```python
#Generate a more complex chart 
f = graphviz.Digraph('Workflow 2')
f.attr('node', shape='box', style='filled', color='lightgrey', fixedsize='true', width='1.8')
f.node('Modeling')
f.node('Tabular data')
f.node('MRI images')

f.attr('node', shape='box', fixedsize='true', width='1.4')
f.edge('Load data', 'Tabular data')
f.edge('Load data', 'MRI images')
f.edge('Tabular data', 'MRI images', label='IDs')
f.edge('MRI images', 'Modeling', label='Normalization | Train set')
f.edge('Modeling', 'Make predictions', label='Test set')
f.edge('Make predictions', 'Output')
f.edge('Make predictions', 'Error analysis')

#f.view()
```

```python
#Generate a more complex chart 
h = graphviz.Digraph('Disease progression')
h.attr('node', shape='box', style='filled', color='lightgrey', fixedsize='true', width='1.8')

h.edge('Healthy', 'Preclinical AD', label='No symptoms \n possible changes in brain')
h.edge('Preclinical AD', 'MCI', label = 'Very mild symptoms')
h.edge('MCI', 'Mild dementia', label='Some compromised \n everyday activities')
h.edge('Mild dementia', 'Moderate dementia', label='Many compromised \n everyday activities')
h.edge('Moderate dementia', 'Severe dementia', label='Most everyday \n activities compromised')

h.graph_attr['rankdir']='LR'
h.edge_attr.update(arrowhead='normal', arrowsize='1')
h.edge_attr.update(fontsize='10.0')
h.view()
```
